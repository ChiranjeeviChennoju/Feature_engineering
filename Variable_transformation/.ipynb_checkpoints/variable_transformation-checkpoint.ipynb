```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pylab
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("student_performance_updated_1000.csv")
```

```python
NUMERICAL_VARS = ['StudyHoursPerWeek', 'AttendanceRate', 'PreviousGrade', 'FinalGrade']
```

```python
df.info()
```

```python
## Fill missing data with random sample:

def impute_na(data, variable):
    # function to fill na with a random sample
    df = data.copy()
    
    # random sampling
    df[variable+'_random'] = df[variable]
    
    # extract the random sample to fill the na
    random_sample = df[variable].dropna().sample(df[variable].isnull().sum(), random_state=0)
    
    # pandas needs to have the same index in order to merge datasets
    random_sample.index = df[df[variable].isnull()].index
    df.loc[df[variable].isnull(), variable+'_random'] = random_sample
    
    return df[variable+'_random']
```

```python
for i in NUMERICAL_VARS:
    df[i] = impute_na(df, i)
```

```python
df.info()
```

#### We have filled the missing values of the numerical columns with random sampling

#### Transforming AttendanceRate

```python
def diagnostic_plots(df, variable):
    # function to plot a histogram and a Q-Q plot
    # side by side, for a certain variable
    
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    df[variable].hist()

    plt.subplot(1, 2, 2)
    stats.probplot(df[variable], dist="norm", plot=pylab)

    plt.show()
    
diagnostic_plots(df, 'AttendanceRate')
```

```python
### Logarithmic transformation
df['AttendanceRate_log'] = np.log(df.AttendanceRate)

diagnostic_plots(df, 'AttendanceRate_log')
```

#### Log is not better

```python
### Reciprocal transformation
df['AttendanceRate_reciprocal'] = 1 / df.AttendanceRate

diagnostic_plots(df, 'AttendanceRate_reciprocal')
```

```python
# Square root transformation

df['AttendanceRate_sqr'] =df.AttendanceRate**(1/2)

diagnostic_plots(df, 'AttendanceRate_sqr')
```

```python
# Exponential transformation
df['AttendanceRate_exp'] = df.AttendanceRate**(1/1.2) 

diagnostic_plots(df, 'AttendanceRate_exp')
```

```python
#Box-Cox transformation
df['AttendanceRate_boxcox'], param = stats.boxcox(df.AttendanceRate) 

print('Optimal : ', param)

diagnostic_plots(df, 'AttendanceRate_boxcox')
```

#### Here we can see that the box-coz has given the bell shape ( normal distribution)
