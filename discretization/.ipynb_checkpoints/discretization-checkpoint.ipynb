```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options for better visualization
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# Set style for better plots
plt.style.use('dark_background')
sns.set_palette('husl')
```

```python
print("\nMissing values:", df['StudyHoursPerWeek'].isnull().sum())
```

```python
# Visualize the distribution of StudyHoursPerWeek
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.hist(df['StudyHoursPerWeek'].dropna(), bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Study Hours Per Week', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Study Hours Per Week (Before Discretization)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.boxplot(df['StudyHoursPerWeek'].dropna(), vert=True)
plt.ylabel('Study Hours Per Week', fontsize=12)
plt.title('Boxplot of Study Hours Per Week', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Apply equal width discretization to StudyHoursPerWeek
# Let's create 4 bins: Low, Medium, High, Very High

# Create a copy to work with (excluding missing values for now)
study_hours_clean = df['StudyHoursPerWeek'].dropna()

# Calculate bin details
min_val = study_hours_clean.min()
max_val = study_hours_clean.max()
num_bins = 4
width = (max_val - min_val) / num_bins

print("Equal Width Discretization Details:")
print(f"Minimum value: {min_val}")
print(f"Maximum value: {max_val}")
print(f"Number of bins: {num_bins}")
print(f"Width per bin: {width:.2f}")

# Apply pd.cut() with labels
study_hours_binned = pd.cut(study_hours_clean, 
                             bins=num_bins, 
                             labels=['Low', 'Medium', 'High', 'Very High'])

print("\n" + "="*60)
print("\nValue counts in each bin:")
print(study_hours_binned.value_counts().sort_index())

print("\nPercentage distribution:")
print(study_hours_binned.value_counts(normalize=True).sort_index() * 100)
```

```python
# Visualize the discretized variable
plt.figure(figsize=(12, 6))

study_hours_binned.value_counts().sort_index().plot(kind='bar', color='steelblue', edgecolor='black')
plt.xlabel('Study Hours Category (Equal Width Bins)', fontsize=12, fontweight='bold')
plt.ylabel('Number of Students', fontsize=12, fontweight='bold')
plt.title('Equal Width Discretization of Study Hours Per Week', fontsize=14, fontweight='bold')
plt.xticks(rotation=0)
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for i, v in enumerate(study_hours_binned.value_counts().sort_index()):
    plt.text(i, v + 5, str(v), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("\nObservation: Notice how bins have equal width but UNEQUAL frequency (number of students)")
```

Let's apply equal frequency discretization to **PreviousGrade** variable.

```python
# Check the PreviousGrade variable
print("PreviousGrade - Basic Statistics:")
print(df['PreviousGrade'].describe())

print("\nMissing values:", df['PreviousGrade'].isnull().sum())
```

```python
# Visualize the distribution of PreviousGrade
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.hist(df['PreviousGrade'].dropna(), bins=30, edgecolor='black', alpha=0.7, color='orange')
plt.xlabel('Previous Grade', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Previous Grade (Before Discretization)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.boxplot(df['PreviousGrade'].dropna(), vert=True)
plt.ylabel('Previous Grade', fontsize=12)
plt.title('Boxplot of Previous Grade', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Apply equal frequency discretization to PreviousGrade
# Let's create 4 bins: Poor, Fair, Good, Excellent

# Create a copy to work with (excluding missing values)
prev_grade_clean = df['PreviousGrade'].dropna()

# Apply pd.qcut() with labels
prev_grade_binned = pd.qcut(prev_grade_clean, 
                            q=4, 
                            labels=['Poor', 'Fair', 'Good', 'Excellent'])

print("Equal Frequency Discretization Details:")
print(f"Total observations: {len(prev_grade_clean)}")
print(f"Number of bins: 4")
print(f"Expected frequency per bin: ~{len(prev_grade_clean) / 4:.0f} students")

print("\n" + "="*60)
print("\nValue counts in each bin:")
print(prev_grade_binned.value_counts().sort_index())

print("\nPercentage distribution:")
print(prev_grade_binned.value_counts(normalize=True).sort_index() * 100)
```

```python
# Let's see the actual bin edges (quantile boundaries)
print("Quantile boundaries (bin edges):")
quantiles = prev_grade_clean.quantile([0, 0.25, 0.5, 0.75, 1.0])
print(quantiles)

print("\nBin ranges:")
print(f"Poor:      [{quantiles[0.00]:.1f}, {quantiles[0.25]:.1f}]")
print(f"Fair:      ({quantiles[0.25]:.1f}, {quantiles[0.50]:.1f}]")
print(f"Good:      ({quantiles[0.50]:.1f}, {quantiles[0.75]:.1f}]")
print(f"Excellent: ({quantiles[0.75]:.1f}, {quantiles[1.00]:.1f}]")
```

```python
# Visualize the discretized variable
plt.figure(figsize=(12, 6))

prev_grade_binned.value_counts().sort_index().plot(kind='bar', color='coral', edgecolor='black')
plt.xlabel('Previous Grade Category (Equal Frequency Bins)', fontsize=12, fontweight='bold')
plt.ylabel('Number of Students', fontsize=12, fontweight='bold')
plt.title('Equal Frequency Discretization of Previous Grade', fontsize=14, fontweight='bold')
plt.xticks(rotation=0)
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for i, v in enumerate(prev_grade_binned.value_counts().sort_index()):
    plt.text(i, v + 2, str(v), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("\nObservation: Notice how bins have EQUAL frequency (number of students) but unequal width!")
```
